```python
from tkinter import *
from tkinter import font as tkFont  
import time
import datetime

root = Tk()
root.config(bg='orange')
root.title('2048')
root.geometry('400x600')

start = time.time()
pixel = PhotoImage(width=100, height=100)
c = 0

helv18 = tkFont.Font(family='Helvetica', size=18, weight='bold')

janela = Label(root, bg = 'orange')
janela.grid(rows = 5, columns = 5)

def reset():
    pass
    
    
btn = Button(janela,text='RESET', compound="center", image=pixel, command=reset(), height=40, width=80)
btn['font'] = helv18
btn.grid(row=0,column=0, padx=5, pady=5)

text = Text(janela, height = 1, font=helv18)
text.grid(row=0, column=2, padx=5, pady=5)

pontos = Label(text, height = 1, font=helv18, text='PONTOS:         ')
pontos.grid(row=0, column=3, padx=5, pady=5)

times = Label(janela,text='', compound="center", image=pixel, height = 40,width=80,font=helv18)
times.grid(row=0, column=1, padx=5, pady=5)

velha = Label(janela)
velha.grid(rows=3, columns = 3)


def clock():
    global c, start
    tai = time.time()
    t = tai - start
    if t!='':
        times.config(text=datetime.timedelta(seconds=round(t)))
    times.after(1000,clock)
    
    
clock()
root.mainloop()
```

```python
import tkinter as tk
from tkinter import font as tkFont  

def bigger():
    size = font.cget("size")
    font.configure(size=size+2)

def smaller():
    size = font.cget("size")
    size = max(2, size-2)
    font.configure(size=size)

root = tk.Tk()
font = tkFont.Font(family="Helvetica", size=12)

toolbar = tk.Frame(root)
container = tk.Frame(root)

toolbar.pack(side="top", fill="x")
container.pack(side="top", fill="both", expand=True)

bigger = tk.Button(toolbar, text="Bigger", command=bigger)
smaller = tk.Button(toolbar, text="Smaller", command=smaller)

bigger.pack(side="left")
smaller.pack(side="left")

pixel = tk.PhotoImage(width=1, height=1)
for row in range(3):
    container.grid_rowconfigure(row, weight=1)
    for column in range(3):
        container.grid_columnconfigure(column, weight=1)
        button = tk.Button(container, font=font, text="x",
            image=pixel, , width=20, height=20)
        button.grid(row=row, column=column)

root.mainloop()
```
