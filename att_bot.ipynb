```python
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
from time import sleep
import pyautogui as pag
import pyperclip as pc
import keyboard as kb


wb = load_workbook('C:/Users/pdv/Desktop/plan.xlsx')
plan = wb['forns']

def pega_forn(ranger):
    global plan
    for i in range(2,ranger):
        forn = plan[f'A{i}'].value
        if not forn:
            break
        yield forn,i
        
def cola(p):
    kb.press('ctrl')
    sleep(0.2)
    kb.press(p)
    sleep(0.2)
    kb.release('ctrl')
    kb.release(p)   
    
    
pag.hotkey('ctrl','shift','capslock')    
sleep(1)
for i,n in pega_forn(400):
    pag.click(816,238)
    print(i[-6:])
    pc.copy(i[-6:])
    pag.click(918,538)
    cola('v')
    kb.press_and_release('enter')
    sleep(1)
    pag.moveTo(737,491)
    pag.mouseDown(button='left')
    pag.moveTo(576,493,0.5)
    pag.mouseUp(button='left')
    cola('c')
    cnpj = pc.paste()
    plan[f'b{n}'] = cnpj
    
    
wb.save(filename = 'C:/Users/pdv/Desktop/plan.xlsx' )
```
